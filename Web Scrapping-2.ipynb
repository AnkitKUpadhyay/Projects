```python
import pandas as pd

# Webpage url                                                                                                               
url = 'https://en.wikipedia.org/wiki/List_of_Nobel_laureates_in_Physics'

# Extract tables and print the column containing Year and name of Laureate
def extract_wikipedia_table():
    read_url_table = pd.read_html(url)
    url_first_table = read_url_table[0]  # Get first table
    global new_table
    new_table = url_first_table[['Year','Laureate[A]']]
    laureate_table = url_first_table[['Laureate[A]']]  # print(df3)
    laureate_dictionary=laureate_table.to_dict('dict') # print(laureate)
    for x in laureate_dictionary:
        laureate_2=laureate_dictionary[x]
    global lau,final_list 
    lau=[]
    final_list=[]
    for x in laureate_2:
        name=laureate_2[x]
        lau.append(name)


   

# printing two lines of wikipedia on laureate and storing in an array  
def extract_wikipedia_lines():
    global about
    about=[]
    import wikipedia
    
    for x in lau:
        except_search = wikipedia.search(x, results = 5) 
        first_item= except_search[0]
        final_list.append(first_item)
        for i in final_list:
            try:
                result = wikipedia.summary(i, sentences = 2)
                about.append(result)
            except Exception:
                new_search=wikipedia.search(i, results = 5)
                second_item= new_search[1]
                result = wikipedia.summary(second_item, sentences = 2)
                about.append(result)
        

def nobel_prize_table():
    from pandas import DataFrame
    description_column = DataFrame (about,columns=['description'])  # print (df4)
    horizontal_stack = pd.concat([new_table, description_column], axis=1)
    horizontal_stack
    horizontal_stack.to_csv (r'C:\Users\ankit\OneDrive\Desktop\nobel_laureate.csv', index = False, header=True)




extract_wikipedia_table()
extract_wikipedia_lines()
nobel_prize_table()



```
