```python
## Asking for total number of elements

n= int(input("Enter the number of elements: "))
X=[]
Y=[]
global sumx, sumy, product_sum, x_square_sum, a, b, z
sumx=0
sumy=0
product_sum=0
productxy=[]
X2=[]
x_square_sum=0

##calculating the sum of x and y values

for i in range(n):
    Xi= float(input())
    Yi= float(input())
    sumx=sumx+Xi
    sumy=sumy+Yi
    product_sum=product_sum+ Xi*Yi
    product=Xi*Yi
    x_square=Xi*Xi
    x_square_sum=x_square_sum + x_square
    X.append(Xi)
    Y.append(Yi)
    productxy.append(product)
    X2.append(x_square)

##preparing the table with X and Y values

import pandas as pd
from pandas import DataFrame
df = pd.DataFrame(list(zip(X, Y, X2, productxy)),columns =['X', 'Y', 'X2', 'xy'])


## defining normal equations and printing them

def normal_equations():
    print("The normal equations are: ")
    print( str(sumx) + 'a' +' ' + '+' + ' ' + str(n) + 'b' + ' ' + '=' + ' ' + str(sumy))
    print( str(x_square_sum) + 'a' +' ' + '+' + ' ' + str(sumx) + 'b' + ' ' + '=' + ' ' + str(product_sum))
    print("\n")
    

##Calculating values of a and b:

a=((n*(-product_sum))-(sumx*(-sumy)))/((sumx*sumx)-(x_square_sum*n))
b=(((-sumy)*x_square_sum)-((-product_sum)*sumx))/((sumx*sumx)-(x_square_sum*n))


## drawing the graph

def solution():
    print("The curve which fits the points is: ")
    print("y = " + str(a) + 'x' + ' '+ '+'+ ' ' + str(b) )
    print("\n")
    import matplotlib.pyplot as plt
    import numpy as np
    x = np.linspace(0,max(X)+5,100)
    y = a*x+b
    plt.xlim([0, max(X)+5.0])
    plt.ylim([0, max(Y)+5.0])
    plt.plot(x, y, '-r', label='Graph of' + ' y= '+ str(a)+'x'+ '+'+ str(b))
    plt.scatter(X, Y)
    for i_x, i_y in zip(X, Y):
        plt.text(i_x, i_y, '({}, {})'.format(i_x, i_y))
    plt.title('Graph of' + ' y= '+ str(a)+'x'+ '+'+ str(b))
    plt.xlabel('x-axis', color='#006400')
    plt.ylabel('y-axis', color='#006400')
    plt.legend(loc='upper left')
    plt.grid()
    plt.show()
    
    
def evaluate():
    c=int(input("Enter the value of x: "))
    z=a*c+b
    print(z)

          
##displaying the output
print("\n\n***************************************************************************************\n\n")
display(df)
print("\n")
normal_equations()
solution()
print("\n")
evaluate()



```
